# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pandas import DataFrame
```

```python
data = {'names':['steve', 'john', 'richard', 'sarah', 'randy', 'micheal', 'julie'],
        'age':[20, 22, 20, 21, 24, 23, 22],
        'gender':['Male', 'Male', 'Male', 'Female', 'Male', 'Male', 'Female'],
        'rank':[2, 1, 4, 5, 3, 7, 6]}

df=DataFrame(data) # create dataframe
df
```

### Bar Chart - Matplotlib and Seaborn

```python
# Bar Chart in Matplotbli
plt.bar(df['names'], df['age']) # x-axis = names, y-axis= age
plt.xlabel('Names') # label for x axis
plt.ylabel('Age') # label for y axis
plt.title('Comparing Ages')
plt.show() # Without plt.show(), in some environments, especially scripts or non-interactive terminals, the plot wonâ€™t appear at all.
```

```python
# Bar Chart in Seaborn
plot = sns.barplot(data=df, x='names', y = 'age')
plot.set_title("Comparing Ages")
plt.show()
```

## Line Chart

```python
# Matploblib
plt.plot(df['names'], df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing Ages')
plt.show()
```

```python
# Seaborn
plot = sns.lineplot(data=df, x='names', y = 'age')
plt.show()
```

## Pie Chart

```python
# Matplotlib
plt.pie(df['age'], labels = df['names'])
plt.title("Age Comparison")
plt.show()
```

```python
# Seaborn
colors = sns.color_palette('pastel')[0:7]
plt.pie(df['age'], labels = df['names'], colors = colors)
plt.show()
```

## Define Elements of a Plot - Object Oriented Plotting

### There are 4 steps in object oriented plotting
1. Create a blank figure object
2. Add axes to the figure
3. Generate plots within the figure oject
4. Specify plotting and layout parameters for the plot(s) within your figure

```python
from numpy.random import randn
from matplotlib import rcParams
```

```python
%matplotlib inline  
# Itâ€™s a magic command in Jupyter Notebooks (not standard Python). 
# Tells Jupyter to - Render matplotlib plots directly below the code cell where the plot is generated
```

```python
rcParams['figure.figsize']= 5,4

x = range(1,10) # x is a range from 1 to 9 (Python excludes the last value of range(1, 10)).
y = [1,2,3,4,0,4,3,2,1] # y is a list of values that will be plotted on the Y-axis.

fig = plt.figure() # Creates a new blank figure â€” think of it as a canvas.
ax = fig.add_axes([.1,.1,.8,.8]) # Adds an Axes object. defines the position and size. Format: [left, bottom, width, height]

ax.plot(x,y)
```

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure()
ax = fig.add_axes([.1,.1,.8,.8])

ax.set_xlim([1,9]) # Limits the x-axis to values between 1 and 9. Anything outside this range is not shown.
ax.set_ylim([0,5]) # Limits the y-axis to values between 0 and 5.

ax.set_xticks([0,1,2,3,4,5,6,8,9,10]) # Specifies exact positions where ticks appear on the x-axis
ax.set_yticks([0,1,2,3,4,5]) # Specifies ticks on the y-axis at these values.

ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,.8,.8])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.grid()
ax.plot(x,y)
```

### Generating multiple plots in one figure with subplots 

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1,2) # Creates 1 row and 2 columns of subplots (axes).
ax1.plot(x) # plots the values of x against their indices (0 to 8).
ax2.plot(x,y) # plots y against x.
plt.show()
```
